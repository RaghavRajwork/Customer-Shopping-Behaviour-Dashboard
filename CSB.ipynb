```python
import pandas as pd
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
median_table = df.groupby('Category')['Review Rating'].median()
print(median_table)
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns= {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

create new column for age group


```python
labels = ['Young-Adult', "Adult" , 'Middle_aged', 'Senior']
df['age_group'] = pd.qcut(df['age'] , q = 4 , labels=labels)
```

```python
df[['age','age_group']].head(10)
```

create new column for purchase frequency days

```python
frequency_map = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    "Quarterly" : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}

df['purchase_freq_days'] = df['frequency_of_purchases'].map(frequency_map)
```

```python
df[['customer_id' , 'frequency_of_purchases' , 'purchase_freq_days']].head(10)
```
